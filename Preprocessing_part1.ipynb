```python
import pandas as pd
df = pd.read_csv('./전처리전데이터/오염+날씨+녹지+차+인구.csv', index_col = 0)
df
```

```python
#미사용 컬럼 삭제
df = df.drop(columns=['key_0', '측정소코드', '측정소명', '주소', '지점', '지점명','구분별(1)','구분별(2)','시점','일시'], axis=1)
```

```python
#이상한 자동차 컬럼 삭제('등유','알코올','태양열','LNG','하이브리드(LNG+전기)')
df = df.drop(columns=['등유','알코올','태양열','LNG','하이브리드(LNG+전기)'], axis=1)
```

```python
# 전체 데이터 대비 결측치 퍼센트를 계산
total_rows = len(df)
missing_percentages = {}

for column in df.columns:
    missing_count = df[column].isnull().sum()
    missing_percentage = (missing_count / total_rows) * 100
    if missing_percentage >= 10:
        missing_percentages[column] = missing_percentage

# 결측치 비율이 10% 이상인 컬럼 출력
for column, percentage in missing_percentages.items():
    print(f"Column {column} has {percentage:.2f}% missing values")
```

```python
# 결측이 10% 이상인 컬럼 삭제(각각 96.7%,99.73%,44.33%,49.24%,100%,70.51%)
df = df.drop(columns=['적설(cm)','3시간신적설(cm)','운형(운형약어)','최저운고(100m )','지면상태(지면상태코드)','현상번호(국내식)'], axis=1)
```

```python
#QC 플래그 전처리 ( QC플래그 열은 결측과 오측을 검사하는 컬럼이기 때문에 삭제 ) 
df = df.drop(columns=['기온 QC플래그','강수량 QC플래그','풍속 QC플래그','풍향 QC플래그','습도 QC플래그','현지기압 QC플래그','해면기압 QC플래그','일조 QC플래그','일사 QC플래그','지면온도 QC플래그'], axis=1)
```

```python
# 휘발유와 경유 콤마 제거 후 숫자로 변환

df['휘발유'] = df['휘발유'].apply(lambda x: float(x.replace(',', '')) if isinstance(x, str) and ',' in x else x).astype(float)
df['경유'] = df['경유'].apply(lambda x: float(x.replace(',', '')) if isinstance(x, str) and ',' in x else x).astype(float)
```

```python
# 수정하고자 하는 칼럼 목록
columns_c = ['휘발유', '경유', '엘피지', '전기', 'CNG',
                     '하이브리드(휘발유+전기)', '하이브리드(경유+전기)', '하이브리드(LPG+전기)',
                     '하이브리드(CNG+전기)', '수소', '기타연료']
```

```python
# 각 칼럼의 값을 0.0484를 곱한 값으로 변경 -> 노원 자동차 보유율로 보정

for column in columns_c:
    # 숫자로 변경 안되는 값들 NAN로 설정
    #df[column] = pd.to_numeric(df[column], errors='coerce')
    # NaN이 아닌 값만 0.0484를 곱함
    df[column] = df[column].apply(lambda x: int(x*0.0484) if pd.notnull(x) else x)
```

```python
#컬럼명 보정
df = df.rename(columns={'엘피지': 'LPG'})
df = df.rename(columns={'PM25': 'PM2.5'})
```

```python
df.info()
```

```python
#추가 전처리1 - dtype 변경
#df['수벽(개소)']' object  => int64 변경
df['수벽(개소)'].unique() #array(['1', '-'], dtype=object)

# '-'값 0으로 치환
df['수벽(개소)'] = df['수벽(개소)'].replace('-', '0')

# 열의 데이터 타입을 int64로 변경
df['수벽(개소)'] = df['수벽(개소)'].astype('int64')
df['수벽(개소)']
```

```python
#추가 전처리2 - dtype 변경
#df['수벽(면적)'] object  => int 변경
df['수벽(면적)'].unique() #array(['12000.0', '-'], dtype=object)

# '-'값 0으로 치환
df['수벽(면적)'] = df['수벽(면적)'].replace('-','0')
# '12000.0'값 '12000'으로 치환
df['수벽(면적)'] = df['수벽(면적)'].replace('12000.0','12000')

# 열의 데이터 타입을 int64로 변경
df['수벽(면적)'] = df['수벽(면적)'].astype('int64')

df['수벽(면적)']
```

```python
#추가 전처리3 - dtype 변경 & 데이터쉼표분리제거
#df['인구'] object => int64
df['인구'].unique() #숫자 데이터 쉼표로 이어짐 확인

#쉼표 제거
df['인구'] = df['인구'].str.replace(',', '')

# 열의 데이터 타입을 int64로 변경
df['인구'] = df['인구'].astype('int64')

df['인구']
```

```python
#추가 전처리4 - dtype 변경
#float형일 필요 없는 xxx.0데이터만을 취하는 컬럼 -> int64형으로 변경 (메모리 관점)

#float형 컬럼명 모두 추출
float_col = df.select_dtypes(include='float').columns.tolist()

# 소숫점 이하가 0인 경우 int형으로 변환
for col in float_col:
    if all(df[col] % 1 == 0):
        df[col] = df[col].astype('int64')
    
#float64 -> int64 대상 컬럼
#df['PM10'], df['PM2.5'], df['풍향(16방위)'], df['전운량(10분위)'], df['중하층운량(10분위)'], df['최저운고(100m)'],
#df['총합계(면적)'], df['시설녹지(면적)'], df['일반녹지(면적)'], df['분리대(면적)'], df['수림대(면적)'], 
#df['하천변조경(면적)'], df['간이휴게소(면적)'], df['지하철환기구주변(면적)'], df['건물주변(면적)'], 
#df['아파트 및 학교(면적)'], df['침수공간조성(면적)'], df['기타(면적)'], df['휘발유'], df['경유']
df.info()
```

```python
#기타면적 없는 값 전처리하기
df['기타(면적)'].unique()    #-> array(['-', '100', 100, 5647], dtype=object)

#'100' 인 값 처리하기 object ->float
df.loc[df['기타(면적)'] == '100', '기타(면적)'] = 100

# 기타(면적)의 '-'값은 2018년도 데이터이고 2018-2021년까지는 기타(개소)의 값이 1이다. 따라서 기타(면적)에서 '-' 값을 100으로 치환한다
df.groupby(['기타(개소)', '기타(면적)']).size().reset_index().rename(columns={0:'count'})
```

```python
#'-' 인 값 처리하기 object ->float
df.loc[df['기타(면적)'] == '-', '기타(면적)'] = 100
df['기타(면적)'] = df['기타(면적)'].astype(int)
```

```python
df.to_csv('./전처리전데이터/raw.csv', index=False, encoding="utf-8-sig")
```
